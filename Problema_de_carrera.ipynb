```python
import numpy as np
import matplotlib.pyplot as plt
print("vamos a resulver la siguiente ecuación diferencial dtdy​=3y+2,y(0)=1")

def f(t, y):
    return 3 * y + 2

t0 = 0
tf = 2
y0 = 1
h = 0.1

t_values = np.arange(t0, tf + h, h)
y_values = np.zeros(len(t_values))
y_values[0] = y0

for i in range(1, len(t_values)):
    y_values[i] = y_values[i - 1] + h * f(t_values[i - 1], y_values[i - 1])

plt.plot(t_values, y_values, label="Euler (paso = 0.1)", marker='o')
plt.xlabel("t")
plt.ylabel("y(t)")
plt.title("Solución de EDO: dy/dt = 3y + 2 (Euler)")
plt.grid(True)
plt.legend()
plt.show()
```
