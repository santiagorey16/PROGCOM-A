En el siguiente paso aÃ±adiremos mÃ¡s elementos a la ventana para que el usuario pueda interactuar y el juego empiece a funcionar. Si te gustÃ³, no olvides dar â€œme gustaâ€ y suscribirte para no perderte los prÃ³ximos avances. Â¡Hasta pronto!

```python
import tkinter as tk
import random

# ------------------ Datos del juego ------------------
opciones = ["Piedra", "Papel", "Tijera", "Lagarto", "Spock"]

# Reglas: cada jugada vence a las de su lista
reglas = {
    "Tijera": ["Papel", "Lagarto"],
    "Papel": ["Piedra", "Spock"],
    "Piedra": ["Tijera", "Lagarto"],
    "Lagarto": ["Spock", "Papel"],
    "Spock": ["Tijera", "Piedra"]
}

marcador = {"Jugador": 0, "PC": 0, "Empates": 0}

# ------------------ LÃ³gica ------------------
def jugar(eleccion_jugador: str) -> None:
    """Gestiona una ronda completa."""
    eleccion_pc = random.choice(opciones)

    if eleccion_jugador == eleccion_pc:
        resultado = "ðŸ¤ Â¡Empate!"
        marcador["Empates"] += 1
    elif eleccion_pc in reglas[eleccion_jugador]:
        resultado = f"âœ… Ganaste: {eleccion_jugador} vence a {eleccion_pc}"
        marcador["Jugador"] += 1
    else:
        resultado = f"âŒ Perdiste: {eleccion_pc} vence a {eleccion_jugador}"
        marcador["PC"] += 1

    etiqueta_resultado.config(
        text=(
            f"TÃº elegiste: {eleccion_jugador}\n"
            f"PC eligiÃ³: {eleccion_pc}\n\n"
            f"{resultado}"
        )
    )
    etiqueta_marcador.config(
        text=f"Marcador â€” TÃº: {marcador['Jugador']}  PC: {marcador['PC']}  Empates: {marcador['Empates']}"
    )

def reiniciar() -> None:
    for k in marcador:
        marcador[k] = 0
    etiqueta_resultado.config(text="Haz tu jugada para comenzar.")
    etiqueta_marcador.config(text="Marcador â€” TÃº: 0  PC: 0  Empates: 0")

# ------------------ Interfaz ------------------
ventana = tk.Tk()
ventana.title("Piedra, Papel, Tijera, Lagarto, Spock")
ventana.geometry("560x520")
ventana.configure(bg="#f7f7fb")

titulo = tk.Label(
    ventana,
    text="Piedra, Papel, Tijera, Lagarto, Spock",
    font=("Arial", 16, "bold"),
    bg="#f7f7fb",
)
titulo.pack(pady=12)

marco_botones = tk.Frame(ventana, bg="#f7f7fb")
marco_botones.pack(pady=6)

for opcion in opciones:
    tk.Button(
        marco_botones,
        text=opcion,
        width=16,
        height=2,
        command=lambda o=opcion: jugar(o)
    ).pack(pady=6)

etiqueta_marcador = tk.Label(
    ventana,
    text="Marcador â€” TÃº: 0  PC: 0  Empates: 0",
    font=("Arial", 12, "bold"),
    bg="#f7f7fb",
)
etiqueta_marcador.pack(pady=10)

etiqueta_resultado = tk.Label(
    ventana,
    text="Haz tu jugada para comenzar.",
    font=("Arial", 12),
    bg="#f7f7fb",
    wraplength=520,
    justify="center",
    anchor="center",
)
etiqueta_resultado.pack(pady=20, fill="both", expand=True)

marco_inferior = tk.Frame(ventana, bg="#f7f7fb")
marco_inferior.pack(pady=10)

tk.Button(marco_inferior, text="Reiniciar", command=reiniciar, width=14).grid(row=0, column=0, padx=8)
tk.Button(marco_inferior, text="Salir", command=ventana.destroy, width=14).grid(row=0, column=1, padx=8)

ventana.mainloop()
```
